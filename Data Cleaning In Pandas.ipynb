# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel("C:/Users/nicks/OneDrive/DATA ANALYST/DATASETS/Customer Call List.xlsx")
df
```

## Drop Duplicates

```python
df=df.drop_duplicates()
df
```

## Remove Columns

```python
# Remove columns we dont need
df=df.drop(columns = "Not_Useful_Column") # inplace=true will save it as well
df
```

## Fix Last_Name Column

```python
# Now go column by column and start fixing
# Last_Name column!!

#lstrip takes from the left side, rstrip takes from the right, srtrip takes from both. You can specify those values too.
# by default it only strips white values or spaces    
#df["Last_Name"].str.strip()
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123./")
df
```

## Format Phone_Number Column

```python
# now for the phone numbers
# we want all of them to look like 123-456-7890
# we will get rid of all non numeric values
# replace ^any character except a-z A-Z 0-9 with nothing ('')
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
df
```

```python
# now we format the phone numbers
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# before we apply that formatting we need to change the column to all strings. This needs to be done to each value not the entire column.
df["Phone_Number"].apply(lambda x: str(x))
```

```python
# we now apply define that column
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
# now we run the formatting to change the column to strings
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# we apply it
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# Remove the NaN 
df["Phone_Number"].str.replace('nan--','')
```

```python
# apply it to the column
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
```

```python
# now remove the Na
df["Phone_Number"].str.replace('Na--','')
```

```python
# apply it to the column and show data frame
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

## Clean Up Address Column

```python
# Address column needs to be separated into 3 different columns
df["Address"].str.split(',',1, expand=True)
```

```python
# Because index 3 has another comma separater, we will use 2 instead of 1.
df["Address"].str.split(',',2, expand=True)
```

```python
# now we have to specify the column names to append the 0,1,and 2. This is not replacing the numbers but creating different columns
df[["Street_Address","State","Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

## Clean Up Paying Customer Column

```python
# we want to replace the Yes's and the No's with Y or N 
df["Paying Customer"].str.replace('Yes','Y')
```

```python
# apply it to the column
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
# replace the No's and apply to the column
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

## Clean Up Do_Not_Contact Column

```python
# Same process as Paying Customer column
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

## Remove N/a & NaN From Entire DataFrame

```python
# Fill all non values with a blank
df = df.fillna('')
df
```

## Filter DataFrame For Only Phone Numbers To Call

```python

for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y': #if x in the Do_Not_Contact column is equal to Y then drop
        df.drop(x, inplace=True) #inplace=True or it wont take effect
        
df        
```

```python
# now we get rid of rows that have no value in Phone_Number

for x in df.index:
    if df.loc[x,"Phone_Number"] == '': #if x in the Phone_Number column is equal to blank then drop
        df.drop(x, inplace=True) #inplace=True or it wont take effect
        
df
# we could have also written it out like this
# df = df.dropna(subset="Phone_Number"), inplace=True
```

## Reset The Index

```python
df = df.reset_index(drop=True) #drop=True drops origional index and shows only the new index
```
